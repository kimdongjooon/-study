변수명1.dot(변수명2)로 계산할 수 있습니다. 

##행렬 곱 연산

내적(inner product) 또는 scalar product라고 함

```
import numpy as np
v = np.array([[1,2,3], [4,5,6]]) # 2x3
w = np.array([[1,1],[2,2], [3,3]]) # 3x2
## dot 함수 이용
print(v.dot(w)) # v x w = 2x2 행렬

print(w.dot(v)) # w x v = 3x3 행렬
```

##2.15.2 인수와 반환값
함수에 전달할 변수를 인수라고 한다. 
함수의 출력을 반환값이라 한다. 
1. 반환값 여러개 가능 
2. 어떠한 형태든 인수나 변환값을 만들 수 있다

```
def my_func3(D):
  m = np.mean(D)
  s = np.std(D)
  return m,s
D = np.random.randn(100)

my_func3(D)
```

format() 메소드: 문자열 내 지정한 값을 지정한 형식으로 출력

```
data = np.random.randn(100)
data_m, data_std = my_func3(data)
print( 'mean: {0:3.2f}, std:{1:3.2f}'.format(data_m, data_std))
```

```
output = my_func3(data)
print(output)
print(type(output))
print( 'mean: {0:3.2f}, std:{1:3.2f}'.format(output[0], output[1]))
```

## 2차원 그래프 그리기
그래프를 그리기 위해 matplotlib의 pyplot 라이브러리를 import하고, 이를 plt라는 별칭을 만들어 사용합니다. 또한 주피터 노트북에서 그래프를 표시하기위해 %matplotlib in-line 명령을 추가합니다.

```
# 그래프를 그리기위한 사전 작업
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline # 주피터 노트북에 나타내기를 원한다.
```

```
#data 작성
np.random.seed(1)       # 난수를 발생하기위한 초기치 1로 고정
x = np.arange(10)       # 10개의 x좌표 생성 # 0부터 1증가시켜 10개 생성
y = np.random.rand(10)  # 10개의 y좌표 생성 # rand 0에서 1사이의 난수 10개 생성

# 그래프 표시
plt.plot(x,y) # 꺽은선 그래프를 등록 가로좌표와 세로좌표로 점을 찍고 점을 서로 직선으로 이거서 만든 그래프 출력
plt.show() # 그래프 그리기 다른 편집기에도 대응할 수 있도록 사용
```

##3차 함수 그리기

```
# 함수를 정의
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(x):
  return (x-2)*x*(x+2)

print(f(1))
print(f(np.array([1,2,3]))) #리스트로는 계산불가 numpy배열로 여러개의 변수 값 계산가능

plt.plot(np.array([1,2,3]), f(np.array([1,2,3])))
```
