#값과 자료형 그리고 변수

##자료형
**정수**(integer, int), **실수**(float), **문자열**(string, str), **논리값**(boolean, bool) 4 종류이다.
프로그래밍에서는 값의 종류를 **자료형**, **데이터 타입**(data type) 또는 간단히 **타입**(type)이라고 한다.

```
print(type(1))     # 정수
print(type('1'))   # 문자열
print(type("1"))   # 문자열
print(type(True))  # 논리값
print(type(1.0))   # 실수
```

## 변수 

변수는 보다 편하고 이해하기 쉽게 프로그램 실행을 위함에 필수적이다.


값을 직접기재하는것은 반드시 피해야할 프로그래밍의 기본이다.


쓰는 형식은 **변수 = 넣을 내용 ** 이다. 이를 '=' 기준으로 오른쪽에 기입한 내용을 왼쪽변수에 대입한다. 라고 읽으면 된다.






```
name = 'kim'
age = 24
human = True
weight = 71.2 

print(name)
print(age)
print(human)
print(weight)
```

'=' 이를 대입연산자라고 하는데 대입연산자로 변수값을 변경할 수 있다.

```
print(name)
print(age)
print(human)
print(weight)
print('\n')

name = 'kim dj'
age = 25
human = False
weight = 69.9

print(name)
print(age)
print(human)
print(weight)
```

## 표현식과 산술 연산자 그리고 `math` 모듈

###표현식

표현식은 값, 변수, 연산자, 함수등으로 구성된다.

##산술연산자

수치연산에 사용되는 연산자로 +, -, `*`, /, //, %, **등이있다.

```
a = 7
b = 10

print('a+b = ',a+b)   # 덧셈
print('b-a = ',b-a)   # 뺄셈
print('a*b = ',a*b)   # 곱셈
print('b/a = ',b/a)   # 나눗셈
print('b//a = ',b//a) # 몫
print('b%a = ',b%a)   # 나머지
print('b**a = ',b**a) # 승수
```

## math 모듈

위에서 언급한 ``math`` 모듈에는 수학적 계산에 필요한 상수와 함수를 정의하고 있다.
따라서 수학적 계산을 하는 프로그램을 작성할 때는 ``math`` 모듈에 이미 만들어 놓은 것을 이용하는 것이 좋다.


`math` 모듈에는 `pi`, `e`, `inf`, `nan`와 같은 상수 뿐만 아니라 계승, 제곱근, 로그 함수, 지수 함수, 삼각 함수, 쌍곡선 함수 등과 같은 기본적인 수학 함수가 정의되어 있으므로 `math` 모듈을 이용한 수학적 계산에 익숙해져야 한다.
`math` 모듈에 정의된 상수와 함수에 대한 자세한 내용은 [math 모듈 문서](https://docs.python.org/3/library/math.html)를 참조해야 한다.

```
import math
print(math.e)
print(math.inf, -math.inf)
print(math.nan)
```

$\theta = 2$, $x=1.5$일 때 함수 $f(x) = \theta e^{-\theta x}$의 값은 다음과 같이 구할 수 있다.

```
theta = 2
x = 1.5
fx = theta * math.exp(-theta*x)
print(fx)
```

$\mu=100$, $\sigma=15$, $x=120$일 때 함수 $f(x) = \frac{1}{\sqrt{2\pi} \sigma} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$의 값은 다음과 같이 계산할 수 있다.

```
mu = 100
sigma = 15
x = 120

fx = 1/(math.sqrt(2*math.pi)*sigma) * math.exp(-(x - mu)**2 / (2 * sigma**2))
print(fx)
```

위와 같은 방법을 사용하게되면 시각적으로 이해하기 힘들 수 있다.


그래서 변수를 지정하여 이해하기 쉽게 만들수 있다.

```
mu = 100
sigma = 15
x = 120

denominator = math.sqrt(2*math.pi)*sigma
exp_deno = (x - mu)**2
exp_nume = 2 * sigma**2

fx = 1/denominator * math.exp(-exp_deno/exp_nume)


print(fx)
```
