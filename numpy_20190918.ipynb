###복습

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(x):
  return (x-2)*x*(x+2)

x = np.linspace(-3,3,10)
```

```
plt.plot(x,f(x))
plt.show
```

##연습문제
1. plot() 함수 대신에 scatter() 함수사용하여 본다
2. scatter() 함수의 매개변수 c,s에 점의 색깔과 크기를 지정할 수 있다.

```
x = np.linspace(-5,5,100)
fx = f(x)
plt.scatter(x,fx,c ='black' , s = 1) # s = x로하면 x크기에따라 점의 크기가 변하는 걸 볼 수 있다.
plt.show()
```

##3.1.6 그래프 장식하기

그래프에 수학적 기호를 레이블에 주고 싶으면 '\$수학적기호$'로 나타낼 수 있다.



```
# 함수정의
def f2(x, w):
  return (x-w) * x * (x+2)


# x를 정의
x = np.linspace(-3,3,100)


# 차트 묘사
plt.plot(x, f2(x,2), color= 'teal', label = '$w=2$', )
plt.plot(x, f2(x,1), color= 'cornflowerblue', label ='$w=1$')
plt.legend(loc="upper left")                                    # 볌례를 나태낼 수 있다.
plt.ylim(-15,5)
plt.xlim(-4,4)
plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$', rotation=90)                                  # rotation 레이블 회전각도
plt.grid(True)                                                  # 그래프 배경선을 그어준다
plt.show()
```

##그래프의 색상을 바꿀 수 있다.

1. r빨강, b파랑, g녹색, c시안, m마젠타, y노란색, k검정, w흰색 이라는 알파벳문자로 지정할수 있다.


2. 또다른 방법은 black,gray,blue,conflowerblue을 중심으로사용한다.


3. 또한 color = (255,0,0)과같이 0과255까지의 정수로 tuple형으로 RGB지정가능하다.

##그래프를 여러개 보여주기

```
colors = ['black', 'blue', 'red', 'yellow', 'cyan', 'magenta']

plt.figure(figsize=(10,3))
plt.subplots_adjust(wspace=1, hspace=0.5)  # wspace는 좌우 간격, hspace는 상하 간격을 조정한다.
for i in range(6):
  plt.subplot(2,3, i+1)                 # 작업 할 그래프 지정
  plt.title(i+1)                        # 제목
  plt.plot(x,f2(x,i), c = colors[i])    # 그래프 작성
  plt.ylim(-20,20)                      # 세로축 번위 지정
  plt.grid(True)                        # 그래프에 모눈 나태냄
  
  
plt.show()
```
