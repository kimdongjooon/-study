````
#1
import numpy as np
from scipy.stats import norm 
stdnormal = norm(loc = 0, scale =1)  # 평균이 0이고 표준편차가 1인 표준정규분포 생성

x = np.arange(-3, 3, 0.05)           # -3 에서 3까지 0.05간격으로 생성
density = stdnormal.pdf(x)           # 확률 밀도 함수 생성.
print(x)
print(density)
 type(density)
````

````
# 샘플 2x3행렬로 생성 이들을 무한히 생성할수록 평균이 0에 편차가 1에 가까워진다.
stdnormal.rvs(size=(2,3))
````

````
import matplotlib.pyplot as plt
plt.plot(x, density)             # 확률 밀도 함수
plt.show                         # 처음실행할때 안하면 안뜬다.쥬피터에서 아마도?

````

````
cdf = stdnormal.cdf(x)
plt.plot(x,cdf)                    # 누적 분포함수
plt.show()
````

7. scipy.stats.multivariate_normal 클래스를 이용하여 2변량 정규분포의 밀도함수 그래프를 등고선도로 작성하시오

````
import numpy as np
means = np.array([10,20])                                  # 2변량 정규분포의 평균 벡터
cov_mat = np.array([[4,-4], [-4,9]])                      # 2변량 정규분포의 공분산행렬
from scipy.stats import multivariate_normal
bi_normal = multivariate_normal(mean=means, cov=cov_mat)  # 2변량 정규분포 객체 생성

bi_normal.pdf([[[10,12], [10,12]],[[19,19],[21,21]]])     # 2변량 정규분포 밀도 계산
                                                          # 두 변수의 값을 3차원으로 준다.
                                                          # 다음은 두 변수의 값이 각각 (10,19), (12,19), (10,21), (12,21)인 
                                                          # 4 점에서 밀도를 계산한다.
````

````
x1_lb = means[0] - 3 * np.sqrt(cov_mat[0][0])
x1_ub = means[0] + 3 * np.sqrt(cov_mat[0][0])
x2_lb = means[1] - 3 * np.sqrt(cov_mat[1][1])
x2_ub = means[1] + 3 * np.sqrt(cov_mat[1][1])

n = 100
x1 = np.linspace(x1_lb, x1_ub, n)  # 100개의 값
x2 = np.linspace(x2_lb, x2_ub, n)  # 100개의 값

xx1, xx2 = np.meshgrid(x1, x2)     # 100x100 행렬 2차원

# xx1 행렬, xx2 행렬 두개가 겹친 3차원 행렬을 만든다.
pos = np.empty(xx1.shape + (2,))  #3차원
pos[:, :, 0] = xx1                #3차원
pos[:, :, 1] = xx2                #3차원
````

scipy.stats.multivariate_normal 객체의 pdf() 메소드로 밀도를 계산할 때는 변수의 값이 저장된 2차원 배열을 겹친 3차원 배열을 인자로 주어야 한다.

````
pdf = bi_normal.pdf(pos) # 밀도 계산
````

````
plt.pcolor(x1, x2, pdf)
plt.colorbar()
````

````
ax = plt.contour(x1, x2, pdf, levels = 10)
ax.clabel(fmt='%4.2f')
plt.show()
````

````
plt.pcolor(x1,x2,pdf,cmap='Blues')
plt.colorbar()
ax = plt.contour(x1, x2, pdf, levels=10)
ax.clabel(fmt='%4.2f')
plt.show()
````

5. 난수를 1000개 발생하여 평균,분산,표준편차 2.5, 5, 10, 50 95, 97.5백분위 수를 계산하시오

````
from scipy.stats import chi2
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
````

````
#기초작업
from scipy.stats import chi2
df = 20
chidist = chi2(df)
````

````
import numpy as np
x_ib = 0
x_ub = 60
n = 1000
x = np.linspace(x_ib, x_ub, n)
````

````
#1
dens = chidist.pdf(x)
plt.plot(dens)
````

````
#2
prob = chidist.cdf(x)
plt.plot(x,prob)
````

````
#3
percent = np.array([0.025, 0.05, 0.1, 0.5, 0.95, 0.975])
pot1 = chidist.ppf(percent)
print(pot1)
````

````
#4
print(chidist.mean())
print(chidist.var())
print(chidist.std())
````

````
#5
data = chidist.rvs(size = 1000)
print(np.mean(data))
print(np.var(data))
print(np.std(data))
print(np.percentile(data, percent))
````
